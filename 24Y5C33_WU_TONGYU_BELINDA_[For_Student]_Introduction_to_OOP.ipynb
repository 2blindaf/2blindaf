```python
class Student:
  def __init__(self, sid, fname, lname, gen, cls, dob):
    self.__studentID = sid
    self.__firstname = fname
    self.__lastname = lname
    self.__gender = gen
    self.__classname = cls
    self.__date_of_birth = dob

  def __str__(self):
    return f"Info of {self.__studentID}: First name: {self.__firstname}, Last name: {self.__lastname}, Gender: {self.__gender}, Class: {self.__classname}, Date of birth: {self.__date_of_birth}"

  def __del__(self):
    print(f">>Student {self.__studentID} has been destroyed")
    print()

  def calculateAge(self):
    age = 2024 - int(self.__date_of_birth[-4:])
    print(age)

  def get_firstname(self): #this is an "accessor/getter"
    return self.__firstname

  def get_studentID(self):
    return self.__studentID

  def get_lastname(self):
    return self.__lastname

  def get_gender(self):
    return self.__gender

  def get_classname(self):
    return self.__classname

  def get_date_of_birth(self):
    return self.__date_of_birth

  def set_firstname(self, new): #this is a "mutator/setter"
    self.__firstname = new

  def set_lastname(self, new):
    self.__lastname = new

  def set_gender(self, new):
    self.__gender = new

  def set_classname(self, new):
    self.__classname = new

  def set_date_of_birth(self, new):
    self.__date_of_birth = new

  def set_studentID(self, new):
    self.__studentID = new

def main():
  student1 = Student("S10", "Tom", "Goh", "Male", "2D", "30012010")
  student2 = Student("S25", "Mary", "Lee", "Female", "5C41", "25062007")
  student3 = Student("S7", "John", "Tan", "Male", "6C22", "18012006")
  print(student1)
  print(student1.get_studentID())
  print(student1.get_firstname())
  print(student1.get_lastname())
  print(student1.get_gender())
  print(student1.get_classname())
  print(student1.get_date_of_birth())
  student3.set_firstname("Johnie")
  print(student3.get_firstname())
main()
```
